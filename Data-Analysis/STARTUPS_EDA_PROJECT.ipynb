```python
east  = 0
northeast = 0
central = 0
west = 0
south = 0
north = 0

for i in startup['state']:
        if(i in east_zone):
            east += 1 
        if(i in northeast_zone):
            northeast += 1
        if(i in central_zone):
            central += 1
        if(i in west_zone):
            west += 1
        if(i in south_zone):
            south += 1
        if(i in north_zone):
            north += 1
```
